```python
import random

class Bank:
    def __init__(self):
        self.accounts = {}

    def add_account(self, name, acc_number, balance = 0.0):
        self.accounts['name'] = name
        self.accounts[acc_number+'_number'] = acc_number
        self.accounts[acc_number+'_balance'] = 0.0
        return
    
    def deposit(self, amount,acc_number):
        self.accounts[acc_number+'_balance'] += amount
        return self.accounts[acc_number+'_balance']

    def withdraw(self, amount,acc_number):
        if self.accounts[acc_number+'_balance'] - amount >= 0:
            self.accounts[acc_number+'_balance'] -= amount
            return True
        else:
            return False

    def transfer(self, from_name, to_name, amount):
        if self.accounts[from_name+'_balance'] - amount >= 0:
            self.accounts[from_name+'_balance'] -= amount
            for key in self.accounts.keys():
                if("_number" in key and to_name == self.accounts[key]):    
                    self.accounts[to_name+'_balance'] += amount
        else:
            print('Insufficient balance!')
            return
                

                

bank = Bank()

name = input("Enter customer name: ")
acc_number = random.randint(1, 999)
bank.add_account(name, str(acc_number))
print("Account created successfully, account number is : ",acc_number)
```
